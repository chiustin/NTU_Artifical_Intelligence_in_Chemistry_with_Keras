```python
import pandas as pd
df = pd.read_excel('LOL.xlsx')
df.head()
```

```python
Label = df['類型'].values
print(Label)
```

```python
for i in range(Label.shape[0]):
    if Label[i] == '鬥士':
        Label[i] = 0
    elif Label[i] == '法師':
        Label[i] = 1
    elif Label[i] == '刺客':
        Label[i] = 2
    elif Label[i] == '坦克':
        Label[i] = 3
    elif Label[i] == '射手':
        Label[i] = 4
    elif Label[i] == '輔助':
        Label[i] = 5

print(Label)
```

```python
from keras.utils.np_utils import to_categorical
y_train = to_categorical(Label)
print(y_train)
```

```python
X_train = df.drop(columns=['名稱','類型'])

mean = X_train.mean(axis=0)
std  = X_train.std(axis=0)

X_train = (X_train - mean) / std
print(X_train)
```

```python
#InternalError: Blas GEMM launch failed
import tensorflow as tf
from keras.backend.tensorflow_backend import set_session
config = tf.compat.v1.ConfigProto()
config.gpu_options.allocator_type = 'BFC' #A "Best-fit with coalescing" algorithm, simplified from a version of dlmalloc.
config.gpu_options.per_process_gpu_memory_fraction = 0.8
config.gpu_options.allow_growth =True

#set_session(tf.compat.v1.Session(config=config)) 
tf.compat.v1.keras.backend.set_session(tf.compat.v1.Session(config=config))
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(100, input_dim=11, activation='relu'))
model.add(Dense(6, activation='softmax'))
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
history = model.fit(X_train, y_train, epochs=2000, batch_size=128, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt
history_dict = history.history
print(history_dict.keys())

#loss_values = history_dict['acc']
loss_values = history_dict['accuracy']
#val_loss_values = history_dict['val_acc']
val_loss_values = history_dict['val_accuracy']

epochs = range(1, len(loss_values) + 1)

plt.plot(epochs, loss_values, 'b', color='k', label='Training accuracy')
plt.plot(epochs, val_loss_values, 'b', color='b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

# Feature Selection

```python
import pandas as pd
df = pd.read_excel('LOL.xlsx')

df = df[(df['類型'].isin(['鬥士','法師','射手']))]
print(df['類型'])
```

```python
Label = df['類型'].values

for i in range(Label.shape[0]):
    if Label[i] == '鬥士':
        Label[i] = 0
    elif Label[i] == '法師':
        Label[i] = 1
    elif Label[i] == '射手':
        Label[i] = 2
        
print(Label)
```

```python
from keras.utils.np_utils import to_categorical
y_train = to_categorical(Label)
print(y_train)
```

```python
X_train = df.drop(columns=['名稱','魔力回復','魔力提升','生命提升','生命','生命回復','移動速度','物理攻擊','物理防禦','魔法防禦','類型'])

mean = X_train.mean(axis=0)
std  = X_train.std(axis=0)

X_train = (X_train - mean) / std
print(X_train)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(100, input_dim=2, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer="Adam", metrics=["accuracy"])
history = model.fit(X_train, y_train, epochs=2000, batch_size=128, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt
history_dict = history.history
print(history_dict.keys())

#loss_values = history_dict['acc']
loss_values = history_dict['accuracy']
#val_loss_values = history_dict['val_acc']
val_loss_values = history_dict['val_accuracy']

epochs = range(1, len(loss_values) + 1)

plt.plot(epochs, loss_values, 'b', color='k', label='Training accuracy')
plt.plot(epochs, val_loss_values, 'b', color='b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
